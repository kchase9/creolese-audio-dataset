```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor
import torch

device = "cuda:0" if torch.cuda.is_available() else "cpu"

torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32
print(device)
print(torch_dtype)
```

```python
whisper_model_id = "openai/whisper-large-v3"
whisper_processor = AutoProcessor.from_pretrained(whisper_model_id)
haitian_processor = WhisperProcessor.from_pretrained("ZeeshanGeoPk/haitian-speech-to-text")
haitian_model = WhisperForConditionalGeneration.from_pretrained("ZeeshanGeoPk/haitian-speech-to-text")
```

<h1>Let's run Whisper first!</h1>
This version is going to be with the model that has not been fine tuned for Haitian Creole and for the intent of this project is considered the most commonly used model.

```python
model = AutoModelForSpeechSeq2Seq.from_pretrained(
    whisper_model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True
).to(device)
```

```python
from transformers import pipeline

asr_pipeline = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=whisper_processor.tokenizer,
    feature_extractor=whisper_processor.feature_extractor,
    torch_dtype=torch_dtype,
    device=device,
)
```

```python
result = asr_pipeline("/content/drive/MyDrive/Colab Notebooks/FYP Research Artefacts/proof of concept/3.wav", return_timestamps=True)
print(result)
```

<h1>Let's try the Haitian Creole model!

```python
import torchaudio

sample_path = "/content/drive/MyDrive/Colab Notebooks/FYP Research Artefacts/proof of concept/3.wav"
waveform, sample_rate = torchaudio.load(sample_path)
```

```python
if waveform.shape[0] > 1:
    waveform = waveform.mean(dim=0, keepdim=True)

input_features = haitian_processor(waveform.numpy(), sampling_rate=sample_rate, return_tensors="pt").input_features
```

```python
# generate token ids
predicted_ids = haitian_model.generate(input_features)
# decode token ids to text
transcription = haitian_processor.batch_decode(predicted_ids, skip_special_tokens=True)
print(transcription)
```
